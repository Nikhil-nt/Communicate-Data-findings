## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

Let us begin by plotting a scatterplot for showing relationship between 2 quantitative variables i.e the variable of interest ArrDelay with the supporting variable Distance

```python
# plot a regplot to show the relationship between 2 variables Distance and ArrDelay
plt.figure(figsize = [8, 6])
sb.regplot(data=df, x='Distance',y='ArrDelay')
plt.xlabel('Distance')
plt.ylabel('ArrDelay')
plt.title('Scatterplot of Correlation');
```

The line does not show any positive or negative trend, next i will plot a line plot to further investigate relationship between both the variables

```python
# plot a line plot to show the relationship 
bins = np.arange(df['Distance'].min(),df['Distance'].max()+100,100)
nins_e= bins[:-1] + 500
dis_binned = pd.cut(df['Distance'],bins,include_lowest= True)
Arr_mean=df['ArrDelay'].groupby(dis_binned).mean()
Arr_std= df['ArrDelay'].groupby(dis_binned).std()
```

```python
plt.figure(figsize = [8, 6])
plt.errorbar(x=nins_e,y=Arr_mean,yerr= Arr_std)
plt.xlabel('Distance')
plt.ylabel('ArrDelay')
plt.title('Correlation of Arrival delay and distance')
```

The line plot gives the same results as the scatter plot , and it has some missing data in between the line 

I will plot a scatterplot to see the relationshionship between Arrival and departure

```python
# show the relationship between ArrDelay and DepDelay
plt.figure(figsize = [8, 6])
sb.regplot(data=df,x='DepDelay',y='ArrDelay')
plt.xlabel('DepDelay')
plt.ylabel('ArrDelay')
plt.title('Correlation of Arrival delay and departure delay');
```

The plot clearly shows an upward positive trend where the increase in departure delay lead to increase in arrival delay means these variables are positively correlated

Let us explore other methods to investigate the relationship to ensure they are positively correlated

```python
# plot a line plot for further investigation
bins = np.arange(df['DepDelay'].min(),df['ArrDelay'].max()+50,50)
bins_c= bins[:-1] + 10
dis_binned = pd.cut(df['Distance'],bins,include_lowest= True)
Arr_mean=df['ArrDelay'].groupby(dis_binned).mean()
```

```python
plt.figure(figsize = [8, 6])
plt.errorbar(x=bins_c,y=Arr_mean)
plt.xlabel('DepDelay')
plt.ylabel('ArrDelay')
plt.title('Lineplot of Correlation');
```

The beginning of the line shows departure delay with few minutes led to decrease in arrival delay but as the departure delay kept on increasing the same happened with the arrival delay.There were also variations in the arrival delay.

The next plot will be relationship between departure delay and Distance. These are 2 supporting variables

```python
# Investigate the relationship between DepDelay and Distance using regplot
sb.regplot(data=df,x='Distance',y='DepDelay')
plt.xlabel('Distance')
plt.ylabel('DepDelay')
plt.title('Scatterplot of Correlation');
```

The line shows no upper or lower trend which shows there is no relationship between the variables. The departure does not get delayed due to large distance between the cities

```python
# set appropriate bins for the plot
bins = np.arange(df['Distance'].min(),df['Distance'].max()+100,100)
bin_c= bins[:-1]+10
dis_binned = pd.cut(df['Distance'],bins,include_lowest= True)
Arr_mean=df['DepDelay'].groupby(dis_binned).mean()
Arr_std=df['DepDelay'].groupby(dis_binned).std()
```

## Multivariate Exploration

> Create plots of three or more variables to investigate your data even
further. Make sure that your investigations are justified, and follow from

I will begin a multivariate exploration by plotting a scatterplot with positional encodings of size and variables of DepDelay,Distance,ArrDelay

```python
# plot a scatterplot with positional encodings
plt.figure(figsize=[ 10, 7.27])
sb.regplot(data=df, x='Distance',y='ArrDelay',x_jitter=0.4,fit_reg=True
          ,scatter_kws={'s':df['DepDelay']/3})
sizes = [100,200,300]
base_color = sb.color_palette()[0]
legend_obj=[]
for s in sizes:
    legend_obj.append(plt.scatter([], [], s = s, color = base_color))
plt.legend(legend_obj, sizes)
```

Many big circles lie in the beginning of the plot at which are upwards which means DepDelay has a relationship with ArrDelay but no relationship with Distance.

To investigate the data even further i will create a Pairplot plot and this time with 4 variables by adding diverted variable

```python
# create PairGrid plots to see the relationship between 4 variables 
columns= ['ArrDelay','DepDelay','Distance','Diverted']
g= sb.PairGrid(data = df, vars= columns)
g= g.map_offdiag(plt.scatter)
g.map_diag(plt.hist);
```

In this plot we can only see a relationship between departure delay and arrival delay.

I will plot using heatmaps to try different methods in multivariate exploration

```python
# Plot using heatmaps and add colors to the boxes with 2 decimal points
plt.figure(figsize=[ 10, 7.27])
columns= ['ArrDelay','DepDelay','Distance','Diverted']
sb.heatmap(df[columns].corr(), annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0)
plt.title('Heatmap of 4 variables')
```

The DepDelay and ArrDelay shows darker boxes which means there is relationship only between 2 variables. Lighter boxes shows less relationships
